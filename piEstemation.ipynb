```python
import numpy as np

def monte_carlo_pi(num_points):
    inside_circle = 0
    for _ in range(num_points):
        x = np.random.rand()
        y = np.random.rand()
        if x**2 + y**2 <= 1:
            inside_circle += 1
    return 4 * inside_circle / num_points

num_points = 10000
pi_estimate = monte_carlo_pi(num_points)
print("Estimated Pi:", pi_estimate)
```

```python
import matplotlib.pyplot as plt

def plot_monte_carlo_pi(num_points):
    inside_x = []
    inside_y = []
    outside_x = []
    outside_y = []
    for _ in range(num_points):
        x = np.random.rand()
        y = np.random.rand()
        if x**2 + y**2 <= 1:
            inside_x.append(x)
            inside_y.append(y)
        else:
            outside_x.append(x)
            outside_y.append(y)

    plt.figure(figsize=(6,6))
    plt.scatter(inside_x, inside_y, color='blue', marker='.')
    plt.scatter(outside_x, outside_y, color='red', marker='.')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Monte Carlo Simulation for Pi Estimation')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

plot_monte_carlo_pi(num_points)
```
